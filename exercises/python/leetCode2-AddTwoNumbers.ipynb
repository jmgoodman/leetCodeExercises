# My solution

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # guaranteed non-empty lists
        # according to the problem
        # but the Optional type hints suggest otherwise...
        # ...eh whatever, let's trust the problem description
        
        # init the result stack
        resultStack = []
        carryOver   = 0
        while l1 is not None or l2 is not None: # better form than while True
            if l1 is not None:
                val1 = l1.val
                l1   = l1.next
            else:
                val1 = 0 # hmmm not very pythonic of you, Bruder. But the java answer does something similar (albeit with conditional operators), so you're at least doing the right thing
            
            if l2 is not None:
                val2 = l2.val
                l2   = l2.next
            else:
                val2 = 0
                
            thisSum      = val1 + val2 + carryOver
            
            thisPosition = thisSum % 10
            carryOver    = thisSum // 10
            resultStack.append( ListNode(val=thisPosition,next=None) )
            
        if carryOver > 0:
            resultStack.append( ListNode(val=carryOver,next=None) )
        else:
            pass
        
        # now link the nodes
        # I do it like this because as far as I know, I can't just save the pointer to the root node in python
        while len(resultStack) > 1:
            currentNode          = resultStack.pop()
            resultStack[-1].next = currentNode
        return resultStack[0]
```

# LeetCode Java Solution

```python
public ListNode addTwoNumbers(ListNode l1, ListNode l2) {
    ListNode dummyHead = new ListNode(0);
    ListNode p = l1, q = l2, curr = dummyHead;
    int carry = 0;
    while (p != null || q != null) {
        int x = (p != null) ? p.val : 0;
        int y = (q != null) ? q.val : 0;
        int sum = carry + x + y;
        carry = sum / 10;
        curr.next = new ListNode(sum % 10);
        curr = curr.next;
        if (p != null) p = p.next;
        if (q != null) q = q.next;
    }
    if (carry > 0) {
        curr.next = new ListNode(carry);
    }
    return dummyHead.next;
}
```
